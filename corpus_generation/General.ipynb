```python
df = pd.read_excel("Shipibo_Konibo_G.xlsx", "General", skiprows = 9, usecols = "B:G")
df = df.iloc[1:, :]
df = df[df["Oración sin errores ortográficos"].notna()]
```

```python
print("Numero de oraciones: {}".format(df.shape[0]))
```

```python
sentences = df["Oración sin errores ortográficos"].tolist()
corpus = ' '.join(sentences)
words = [w.strip() for w in corpus.split()]
word_count = Counter(words)
unique_words = list(word_count.keys())

print("Número de palabras únicas (Oración sin errores ortográficos): {}".format(len(unique_words)))
```

```python
sentences = df["Oración con errores ortográficos"].tolist()
corpus = ' '.join(sentences)
words = [w.strip() for w in corpus.split()]
word_count = Counter(words)
unique_words = list(word_count.keys())

print("Número de palabras únicas (Oración con errores ortográficos): {}".format(len(unique_words)))
```

```python
errors = df["Tipos de errores"].tolist()
errors = ','.join(errors)
errors = [w.strip() for w in errors.split(',')]
errors_count = Counter(errors)

print("Tipos de errores: {}".format(errors_count))
```

```python
for _, row in df.iterrows():
    sentence = row["Oración sin errores ortográficos"]
    sentence_with_errors = row["Oración con errores ortográficos"]
    error_types = row["Tipos de errores"]
    
    words = sentence.strip().split(' ')
    words_with_errors = sentence_with_errors.strip().split(' ')
    error_types = error_types.strip().split(',')
    
    errors = []
    idx_error_type = 0
    for idx, word in enumerate(words):
        try:
            word_with_posible_error = words_with_errors[idx]            
        except:
            word_with_posible_error = None
            
        try:
            possible_error_type = error_types[idx_error_type]
        except:
            possible_error_type = None
        
        if(word != word_with_posible_error):
            errors.append("{},{},{}".format(word, word_with_posible_error, possible_error_type))
            idx_error_type += 1
        
    print(';'.join(errors))          
```
